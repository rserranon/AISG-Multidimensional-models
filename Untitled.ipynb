```python
import pandas as pd 

clients = ['AA','AM', 'BT', 'VOL', 'VIVA']
ac = ['A319', 'A320', 'A321','737']
station = ['ACA', 'BJX', 'CZM','DGO','GDL']
svc = ['A','B','C','D','E']
index = pd.MultiIndex.from_product([clients, ac])
columns = pd.MultiIndex.from_product([station, svc])
df = pd.DataFrame(index=index, columns=columns)
df.index.names = ['Client','AC']
df.columns.names = ['Station','Service']
df.fillna(1, inplace=True)
df.to_excel('File-Bogar.xlsx')
df
```

```python
# Assuming you already have the DataFrame 'df'
reshaped_df = df.stack(level='Station').reset_index(level=['AC', 'Station'])
reshaped_df = reshaped_df.pivot_table(index=['Client', 'AC', 'Service'], columns='Station', values='Value')

# Now 'Client', 'AC', and 'Service' are in the index, and 'Station' and 'Value' are in the columns
print(reshaped_df)



```
