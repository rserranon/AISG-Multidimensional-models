```python
import pandas as pd 
import numpy as np

clients = ['AA','AM', 'BT', 'VOL', 'VIVA']
ac = ['A319', 'A320', 'A321','737']
station = ['ACA', 'BJX', 'CZM','DGO','GDL']
svc = ['A','B','C','D','E']
index = pd.MultiIndex.from_product([clients, ac])
print(index)
columns = pd.MultiIndex.from_product([station,svc])
content = [i for i in range(len(index)+len(columns))]
tuples = list(zip(index,columns))
print(tuples)
df = pd.DataFrame( index=index, columns=columns)
df.index.names = ['Client','AC']
df.columns.names = ['Station', 'Service']

# Generate random data with the shape of the MultiIndex
random_data = np.random.rand(len(index), len(columns))

# Assign the random data to the DataFrame
df = pd.DataFrame(random_data, index=index, columns=columns)

print(df)


df.fillna(1, inplace=True)
df.to_excel('File-Bogar.xlsx')
df
```
